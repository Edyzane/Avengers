## REGULARIZATION - LASSO (L1)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Ames_Housing_Clean_Data(remove skew).csv')

x=df.drop('SalePrice',axis=1)
y=df['SalePrice']

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
x
```

```python
y**2
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv_model=LassoCV(eps=0.1,n_alphas=100,cv=5)
```

```python
lasso_cv_model.fit(x_train,y_train)
```

```python
lasso_cv_model.alpha_
```

```python
test_pred=lasso_cv_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE=mean_absolute_error(y_test,test_pred)

MAE**2
```

```python
MSE=mean_squared_error(y_test,test_pred)

MSE**2
```

```python
RMSE=np.sqrt(MSE)

RMSE=RMSE**2

RMSE
```

```python
## train performance

train_predictions=lasso_cv_model.predict(x_train)
MAE=mean_absolute_error(y_train,train_predictions)
MAE**2
```

```python
from joblib import dump, load

sale_price_model = load('ames_linear_model.h5')


def return_prediction(model,sample_json):
    
    lot_front = sample_json['Lot Frontage']
    lot_area = sample_json['Lot Area']
    ovr_qual = sample_json['Overall Qual']
    yr_built = sample_json['Year Built']
    yr_remod = sample_json['Year Remod/Add']
    bsmt_fin = sample_json['BsmtFin SF 1']
    t_bsmt_sf = sample_json['Total Bsmt SF']
    first_flr_sf = sample_json['1st Flr SF']
    gr_liv_area = sample_json['Gr Liv Area']
    gr_cars = sample_json['Garage Cars']
    gr_area = sample_json['Garage Area']
        
    
    sale_price = [[lot_front,lot_area,ovr_qual,yr_built,yr_remod,bsmt_fin,
                   t_bsmt_sf,first_flr_sf,gr_liv_area,gr_cars,gr_area]]
    
        
    prediction = model.predict(sale_price)**2
    
        
    prediction = np.round(prediction,2)
        
    return prediction[0]
```

```python
sales_pred = {'Lot Frontage':141,
                 'Lot Area':31770,
                 'Overall Qual':6,
                 'Year Built':1960,
                 'Year Remod/Add':1960,
                 'BsmtFin SF 1':639,
                 'Total Bsmt SF':1080,
                 '1st Flr SF':1656,
                 'Gr Liv Area':1656,
                 'Garage Cars':2.0,
                 'Garage Area':528}
```

```python
prediction = return_prediction(sale_price_model,sales_pred)
```
