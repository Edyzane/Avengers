## ENCODE STRINGS TO INTERGER

Mapping and replace - https://www.geeksforgeeks.org/mapping-external-values-to-dataframe-values-in-pandas/


## 1. Ordinal Encoding

where the item had a relationship between each item

example = excellent,good,average,poor

```python
df.head(3)
```

```python
lotshapper={'Reg':0,'IR1':1,'IR2':2,'IR3':3}

df = df.replace({"Lot Shape":lotshapper})

# create a dictionary (item:numerical value)

# replace the original strings to numerical value

# import the value from the dictionary
```

```python
slope={'Gtl':0,'Mod':1,'Sev':2}

df = df.replace({"Land Slope":slope})
```

```python
yesno={'Y':0,'N':1}

df=df.replace({"Central Air":yesno})
```

```python
fin={'Fin':0,'RFn':1,'Unf':2,'NA':3}

df=df.replace({'Garage Finish':fin})
```

```python
expo={'Gd':0,'Av':1,'Mn':2,'No':3,'NA':4}

df = df.replace({'Bsmt Exposure':expo})
```

```python
qual={'Ex':0,'Gd':1,'TA':2,'Fa':3,'Po':4,'NA':5}

df = df.replace({"Exter Qual":qual})
df = df.replace({"Exter Cond":qual})
df = df.replace({"Bsmt Qual":qual})
df = df.replace({"Bsmt Cond":qual})
df = df.replace({"Heating QC":qual})
df = df.replace({"Kitchen Qual":qual})
df = df.replace({"Fireplace Qu":qual})
df = df.replace({"Garage Qual":qual})
df = df.replace({"Garage Cond":qual})
```
