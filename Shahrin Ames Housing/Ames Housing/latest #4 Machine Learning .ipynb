## GRID SEARCH CV ELASTIC NET

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,50,100],
              'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}
```

```python
from sklearn.model_selection import GridSearchCV

# import GridSearchCV
```

```python
# verbose number a personal preference

grid_model = GridSearchCV(estimator=base_elastic_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5,
                          verbose=2)

#cv value = maximum value can be same number of features
```

```python
grid_model.fit(x_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
 pd.DataFrame(grid_model.cv_results_)
```

```python
test_pred = grid_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_pred)

MAE
```

```python
MSE=mean_squared_error(y_test,test_pred)

MSE
```

```python
RMSE=np.sqrt(MSE)

RMSE
```

```python
#predict on train set

train_pred=grid_model.predict(x_train)

MSE = mean_squared_error(y_train,train_pred)
RMSE= np.sqrt(MSE)

RMSE
```

```python
df['SalePrice'].mean()
```

```python
RMSE*100/df['SalePrice'].mean()
```

```python
4**2
```

7% percent of error percentage for predicting sales price

```python
sns.scatterplot(x=y_test,y=test_res)
plt.axhline(y=0,color='red',linestyle='--');
```

```python
sns.kdeplot(test_res);
```

```python
sns.displot(test_res,bins=20,kde=True);
```

```python
import scipy as sp
```

```python
df=pd.read_csv('Ames_Housing_Clean_Data(remove skew).csv')
```

```python
df
```

```python
x=df.drop('SalePrice',axis=1)
y=df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```
