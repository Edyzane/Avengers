```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
df = pd.read_csv('Ames Housing Data Encoding.csv')
```

```python
df.head(10)
```

```python
X = df.drop('SalePrice',axis=1)
y = df['SalePrice']

# TRAIN TEST SPLIT

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# SCALE DATA

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

**Ridge**

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=100)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
```

```python
100*RMSE/df['SalePrice'].mean()
```

```python
sns.displot(y_pred,bins=15,kde=True)
```

```python
sns.scatterplot(x=y_test,y=y_pred)
plt.axhline(y=0,color='magenta',linestyle='--')
```

```python
fig,ax = plt.subplots(figsize=(5,8),dpi=100)

_=sp.stats.probplot(y_pred,plot=ax)
```

```python
model = Ridge(alpha=1)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
```

```python
100*RMSE/df['SalePrice'].mean()
```
