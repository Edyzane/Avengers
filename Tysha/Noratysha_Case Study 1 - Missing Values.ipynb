```python
def missing_data(df):
    mis_value = df.isnull().sum().sort_values(ascending=False)/len(df)
    mis_value = mis_value[mis_value>0].sort_values()
    return mis_value
```

```python
mis_value = missing_data(df)
```

```python
mis_value
```

```python
sns.barplot(x=mis_value.index,y=mis_value)
plt.xticks(rotation=90);
```

**Drop row that have only 1 missing value**

```python
df = df.dropna(subset = ['Electrical','Garage Area','Garage Cars','Total Bsmt SF'],axis=0)
```

```python
df
```

```python
mis_value = missing_data(df)
```

```python
sns.barplot(x=mis_value.index,y=mis_value)
plt.xticks(rotation=90);
```

**Fill Missing Value in Mas Vnr Features**

```python
Mas_str = ['Mas Vnr Type']
df[Mas_str] = df[Mas_str].fillna("None")
```

```python
Mas_num = ['Mas Vnr Area']
df[Mas_num] = df[Mas_num].fillna(0)
```

```python
mis_value = missing_data(df)
```

```python
sns.barplot(x=mis_value.index,y=mis_value)
plt.xticks(rotation=90);
```
